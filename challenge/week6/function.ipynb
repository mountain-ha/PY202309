```python
import random

def get_number(): 
    return random.randrange(1, 46)

def set_lotto_numbers(): #로또 뽑는 함수
    lotto = []
    while True:
        num = get_number()
        
        if lotto.count(num) == 0: #중복이 없을때 lotto 리스트에 추가하기
            lotto.append(num)
      
         if len(lotto) >= 6: #lotto리스트에 6개의 원소가 있으면 중지
                break
    return lotto

def match_numbers(user_inputs, lotto_numbers): #두 리스트의 교집합인 same_numbers리스트 생성 함수
    same_numbers = []
    for i in user_inputs:
        for j in lotto_numbers:
            if i == j: same_numbers.append(i)
    return same_numbers

def check_prize(same_numbers): #로또 등수 판별함수
    if len(same_numbers) == 6:
        print("1st")
    elif len(same_numbers) == 5:
        print("2st")
    elif len(same_numbers) == 4:
        print("3st")
    elif len(same_numbers) == 3:
        print("4st")
    else:
        print("no prize :(")
num = 0
print("** Start Lotto **")
print("1) Set the lotto number")
lotto_numbers =set_lotto_numbers()

print("2) Set the yser input")
user_inputs = []
while True:                     #로또 입력받기
    num = int(input("Enter %d-number:"%(len(user_inputs)+1)))
    if num<1 or num>46:
        print("No valid number")
        continue
    if user_inputs.count(num) == 0:
        user_inputs.append(num)
    else:
        print("Same number")
        continue
    if len(user_inputs) >=6:
        break

print("3) March the number")  #두 리스트의 교집합 리스트 만들기
same_number =match_numbers(user_inputs, lotto_numbers)

print("4) Check the prize") #로또 등수 판별
check_prize(same_number)

print("5)Sort the lists") #각 리스트의 숫자 정렬
lotto_numbers.sort()
user_inputs.sort()
 
print("===Summary===")   #로또 번호와 내가 쓴 번호 출력
print("Lotto numbers:", *lotto_numbers)
print("Usuer numbers:", *user_inputs)
```
